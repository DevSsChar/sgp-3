```python
from easyfinetune import tune

# This is it - 3 lines to fine-tune an LLM!
result = tune(
    dataset="/content/vastu_training_dataset_complete.jsonl",
    model="auto",  # Automatically picks best model for T4 (likely phi-3-mini)
    method="lora",  # Use LoRA for efficient fine-tuning
    num_epochs=2,  # Quick training (increase to 3-5 for better results)
)

# Check results
if result.success:
    print("\n" + "="*70)
    print("🎉 TRAINING SUCCESSFUL!")
    print("="*70)
    print(f"✅ Model saved to: {result.model_path}")
    print(f"✅ Validation loss: {result.metrics.get('eval_loss', 0):.3f}")
    print(f"✅ Training time: {result.total_time_minutes:.1f} minutes")
    print("\n📁 Output files:")
    for doc_type, path in result.docs.items():
        print(f"   • {doc_type}: {path}")
else:
    print(f"\n❌ Training failed: {result.message}")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch
import json
import os

print("Loading fine-tuned model...")

# Check if training completed successfully
if not result.success:
    print("❌ Training didn't complete successfully. Cannot load model.")
    print(f"Error: {result.message}")
    raise RuntimeError("Training failed - cannot proceed with model testing")

# Get paths from training result
model_path = result.model_path
print(f"📁 Model path: {model_path}")

# Get the actual base model used during training
# Check the config file that was saved
config_path = os.path.join(model_path, "adapter_config.json")
if os.path.exists(config_path):
    with open(config_path, 'r') as f:
        adapter_config = json.load(f)
        base_model_name = adapter_config.get('base_model_name_or_path')
        print(f"🔍 Detected base model: {base_model_name}")
else:
    # Fallback: try to get from result metadata
    base_model_name = result.metadata.get('model_hf_repo_id')
    if not base_model_name:
        print("❌ Cannot determine base model. Check adapter_config.json")
        raise RuntimeError("Base model not found in adapter config")
    print(f"🔍 Using model from training metadata: {base_model_name}")

try:
    print(f"\n📦 Loading base model: {base_model_name}")
    base_model = AutoModelForCausalLM.from_pretrained(
        base_model_name,
        device_map="auto",
        torch_dtype=torch.float16,
        trust_remote_code=True,
    )

    print(f"📦 Loading LoRA adapter from: {model_path}")
    model = PeftModel.from_pretrained(base_model, model_path)
    model.eval()

    print(f"📦 Loading tokenizer")
    tokenizer = AutoTokenizer.from_pretrained(base_model_name, trust_remote_code=True)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    print("\n✅ Model loaded successfully!")

    # # Test prompts
    # test_prompts = [
    #     "Review: This is an amazing product! I love it.",
    #     "Review: Terrible quality. Very disappointed.",
    #     "Review: Not bad, but could be better.",
    # ]
    test_prompts = [
        "Is toilet in northwest good?",
        "I have master bedroom in north",
        "The toilet is in north east, is that good? if not suggest remedies"
    ]

    print("\n" + "="*60)
    print("🧪 TESTING FINE-TUNED MODEL")
    print("="*60)
    print(f"Model: {base_model_name}")
    print(f"Method: LoRA (fine-tuned adapter)")
    print("="*60)

    for i, prompt in enumerate(test_prompts, 1):
        print(f"\n[Test {i}/{len(test_prompts)}]")
        print(f"Prompt: {prompt}")
        print("-" * 60)

        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=250,
                temperature=0.7,
                do_sample=True,
                pad_token_id=tokenizer.eos_token_id,
            )

        generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
        # Show only the new generated text (remove input prompt)
        response = generated[len(prompt):].strip()
        print(f"Response: {response}")

    print("\n" + "="*60)
    print("✅ Testing complete!")
    print("="*60)

except Exception as e:
    print(f"\n❌ Error loading model: {e}")
    print(f"\nDebug Info:")
    print(f"  • Model path: {model_path}")
    print(f"  • Base model: {base_model_name}")
    print(f"  • adapter_config.json exists: {os.path.exists(config_path)}")
    raise
```

```python
from google.colab import files
import shutil
import os
from tqdm import tqdm
import zipfile

print("📦 Preparing model for download...\n")

output_dir = "./easyfinetune_output"
output_zip_path = "/content/easyfinetune_model.zip"

# Calculate total size first
print("📏 Calculating size...")
total_size = 0
file_list = []
for root, dirs, files_in_dir in os.walk(output_dir):
    for file in files_in_dir:
        filepath = os.path.join(root, file)
        file_list.append(filepath)
        total_size += os.path.getsize(filepath)

print(f"📊 Total size: {total_size / (1024*1024):.1f} MB ({len(file_list)} files)")

# Create zip with progress bar
print("\n🗜️  Compressing files...")
with zipfile.ZipFile(output_zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    with tqdm(total=len(file_list), unit='file', desc='Zipping') as pbar:
        for filepath in file_list:
            arcname = os.path.relpath(filepath, os.path.dirname(output_dir))
            zipf.write(filepath, arcname)
            pbar.update(1)

# Check final zip size
zip_size_mb = os.path.getsize(output_zip_path) / (1024*1024)
print(f"\n✅ Model packaged successfully!")
print(f"📦 Compressed size: {zip_size_mb:.1f} MB")
print(f"📁 Location: {output_zip_path}")

# Download with size warning
if zip_size_mb > 200:
    print("\n⚠️  WARNING: File is very large (>200MB)")
    print("   Colab downloads may fail for large files.")
    print("   Consider using Google Drive instead (see next cell).")

    user_confirm = input("\nProceed with download anyway? (y/n): ")
    if user_confirm.lower() != 'y':
        print("❌ Download cancelled. Use Google Drive method below.")
    else:
        print("\n📥 Starting download...")
        try:
            files.download(output_zip_path)
            print("✅ Download started! Check your browser's download folder.")
        except Exception as e:
            print(f"❌ Download failed: {e}")
            print("💡 Try the Google Drive method in the next cell.")
elif zip_size_mb > 100:
    print("\n⚠️  Note: File is large (>100MB). Download may take a while.")
    print("📥 Starting download...")
    try:
        files.download(output_zip_path)
        print("✅ Download started! Check your browser's download folder.")
    except Exception as e:
        print(f"❌ Download failed: {e}")
        print("💡 Try the Google Drive method in the next cell.")
else:
    print("\n📥 Starting download...")
    try:
        files.download(output_zip_path)
        print("✅ Download started! Check your browser's download folder.")
    except Exception as e:
        print(f"❌ Download failed: {e}")
        print("💡 Try the Google Drive method in the next cell.")
```
